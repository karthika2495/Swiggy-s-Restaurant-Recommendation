#Data Preprocessing

```python
clean_data = pd.read_csv("C:/Users/Karthika/Documents/Project/Client_Project/cleaned_data.csv")
clean_data
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
city_encoder = OneHotEncoder(sparse_output=False, handle_unknown="ignore")

city_encoded = city_encoder.fit_transform(clean_data[['city']])

city_encoded

```

```python
city_df = pd.DataFrame(
    city_encoded,
    columns=city_encoder.get_feature_names_out(['city']),
    index=clean_data.index
)
city_df
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

cuisine_split = clean_data['cuisine'].str.split(',')
cuisine_split
```

```python
cuisine_encoder = MultiLabelBinarizer()
cuisine_encoded = cuisine_encoder.fit_transform(cuisine_split)
cuisine_encoded
```

```python
cuisine_df = pd.DataFrame(
    cuisine_encoded,
    columns=cuisine_encoder.classes_,
    index=clean_data.index
)
cuisine_df
```

```python
numerical_df = clean_data[['rating', 'rating_count', 'cost']]
```

```python
encoded_data = pd.concat(
    [numerical_df, city_df, cuisine_df],
    axis=1
)
encoded_data
```

```python
import pickle

encoder_objects = {
    "city_encoder": city_encoder,
    "cuisine_encoder": cuisine_encoder
}

with open("encoder.pkl", "wb") as f:
    pickle.dump(encoder_objects, f)
```

```python
encoded_data.to_csv("encoded_data.csv")
```

#Recommendation Methodology

```python
encoded_data = pd.read_csv("encoded_data.csv", index_col=0)
encoded_data
```

```python
cleaned_data = pd.read_csv("cleaned_data.csv", index_col=0)
cleaned_data
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(encoded_data)
```

```python
import matplotlib.pyplot as plt

inertia = []
K_range = range(2, 11)

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(scaled_data)
    inertia.append(km.inertia_)

plt.plot(K_range, inertia, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```
